```python
#######
# Objective: Create a scatterplot of 1000 random data points.
# x-axis values should come from a normal distribution using
# np.random.randn(1000)
# y-axis values should come from a uniform distribution over [0,1) using
# np.random.rand(1000)
######

# Perform imports here:
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go

np.random.seed(999)
xx = np.random.randn(1000)
yy = np.random.rand(1000)
# Define a data variable

data = []
data.append(
    go.Scatter(x=xx
                    ,y=yy
                   ,mode = 'markers'
              )
)

# Define the layout
layout = go.Layout(
    title = 'x y', # Graph title
    xaxis = dict(title = 'x'), # x-axis label
    yaxis = dict(title = 'y'), # y-axis label
    hovermode ='closest' # handles multiple points landing on the same vertical
)


# Create a fig from data and layout, and plot the fig
fig = go.Figure(data=data, layout=layout)


pyo.init_notebook_mode(connected=True)
pyo.iplot(fig)
```
