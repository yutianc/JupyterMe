```python
import dash
import dash_core_components as dcc
import dash_html_components as html

import pandas as pd
import numpy as np
from sqlalchemy import create_engine # database connection
import datetime as dt
from IPython.display import display

import plotly.plotly as py # interactive graphing
from   plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
from plotly import tools
```

### Create data source and plot

```python
#create Database
disk_engine = create_engine('sqlite:///SQLiteDatabase/jde0911.db') 
dfcomp = pd.read_sql_query('SELECT  公司编码, -1 * SUM(金额) 金额汇总 '
                       'FROM JDE0911BF201810 '
                       'WHERE 1=1 '
                       ' AND  状态 = "S" '
                       ' AND 科目名称 LIKE "%收入%" '
                       ' AND 财务年度 >="10" '
                       ' AND 公司编码 not in ("05310","05320") '
                       'GROUP BY 公司编码 '
                       ' ORDER BY 金额汇总  DESC '
                       #' LIMIT 1 '
                       , disk_engine)

data=[]

for i in np.arange(0, dfcomp['公司编码'].size,  step = 1):
    #df = pd.read_sql_query('SELECT Agency, Descriptor FROM data LIMIT 3', disk_engine)
    df = pd.read_sql_query('SELECT JDE0911BF201810.公司编码, soho_companys.company_name 公司名称, 总账期间, 科目名称, 状态, -1 * SUM(金额) 金额汇总  '
                           ' FROM JDE0911BF201810 , soho_companys '
                           ' WHERE 1=1 '
                         # ' AND  TRIM(合同号) ="14100003" '
                           ' AND  JDE0911BF201810.公司编码 = soho_companys.company_code  '
                           ' AND  状态 = "S" '
                           ' AND 科目名称 LIKE "%收入%" '
                           ' AND 财务年度 >="10" '
                           ' AND 公司编码= "' + dfcomp.iloc[i,:]['公司编码'] + '"  '
                           ' GROUP BY 公司编码, 总账期间, 科目名称, 状态 '
                           ' ORDER BY  公司编码, 总账期间 '
                           , disk_engine)
    #df

    #data.append([go.scatter(x=df.总账期间, y=df.金额汇总)]
    #pdAll['总账期间'] = pd.to_datetime(pdAll['总账期间'], format='%Y%m')
    data.append(go.Bar(x=pd.to_datetime(df.总账期间, format='%Y%m'), y=df.金额汇总 ,  name= df.loc[0,'公司名称'] ))


#收款
dfinc = pd.read_sql_query('SELECT 总账期间,  SUM(金额) 金额汇总  '
                       ' FROM JDE0911BF201810 '
                       ' WHERE 1=1 '
                     # ' AND  TRIM(合同号) ="14100003" '
                       ' AND  状态 = "S" '
                       ' AND (科目名称 LIKE "%现金%" OR  科目名称 LIKE "%银行存款%" ) '
                       ' AND 财务年度 >="10" '
                       ' GROUP BY 总账期间 '
                       ' ORDER BY  总账期间 '
                       , disk_engine)

data.append(go.Scatter(x=pd.to_datetime(dfinc.总账期间, format='%Y%m'), y=dfinc.金额汇总 ,  name= "收款汇总"))

layout = { 'title': '收入/收款分析'
       ,'margin': {'b': 200} # Make the bottom margin a bit bigger to handle the long text
       , 'xaxis': {'title': '期间', 'tickangle': 40 , 'nticks':30, 'tickformat':'%Y-%m' } # Angle the labels a bit
       , 'yaxis': {'title': 'notMoney', 'showticksuffix': 'all', 'nticks':20 } # Set Y axis label , 'ticksuffix':'(百万)'
       , 'barmode':'stack'
       , 'hovermode': 'closest'
    } 

#most_common_complaints = df # used later
fig = go.Figure(data = data, layout = layout)

init_notebook_mode(connected=True)
#iplot(figure_or_data=fig, filename='FileNameofDashbyPlotly.html')
```

### Run Dash Server for plotly

```python
app = dash.Dash()

htmlList = []
htmlList.append(
        dcc.Graph(id='asdfDash'
                ,figure= fig
                 )
)

app.layout = html.Div(htmlList,style={})
    
if __name__ == '__main__':
       app.run_server()
```
